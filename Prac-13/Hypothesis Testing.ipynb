```
import numpy as np
import pandas as pd
import statsmodels.api as sm
import scipy.stats.distributions as dist
```

# Example 1

**The population proportion of Ireland having heart disease is 49%. Are more people suffering from heart disease in the US**

```
heart_dataset = '/content/drive/MyDrive/FDS Practical/heart.csv'

df = pd.read_csv(heart_dataset)
df.head()

```

```
# null hypothesis => p0 = 0.49
# alternative hypothesis => p > 0.49
```

```
# The "target" field refers to the presence of heart disease in the patient. 
# It is integer valued 0 = no disease and 1 = disease.

p_us = len(df[df['target'] == 1])/ len(df) # calculating population proportion
p_us
```

```
# The population proportion of the sample having heart disease is 0.51 or 51%.

# Perform hypothesis test

se = np.sqrt(0.49 * (1 - 0.49) / len(df)) # calculating standard error

be = p_us # best estimate 
he =  0.49 # hypothesized estimate
test_stat = (be - he)/se # calculating test statistics

test_stat
```

```
pvalue = 2 * dist.norm.cdf(-np.abs(test_stat)) # calculating p-value

pvalue
```

Consider the significance level alpha to be 5% or 0.05. A significance level of 5% or less means that there is a probability of 95% or greater that the results are not random.

Here p-value is bigger than our considered significance level of 0.05. So, we cannot reject the null hypothesis. That means there is no significant difference in population proportion having heart disease in Ireland and the US.

# Example 2

**Population proportion of students who have taken portuguese subject and has access to internet is 90%. Do more students who have taken math subject have access to the internet?**

```
# null hypothesis => p0 = 0.90
# alternative hypothesis => p > 0.90
```

```
math_students_dataset = '/content/drive/MyDrive/FDS Practical/student-mat.csv'

df = pd.read_csv(math_students_dataset)
df.head()
```

```
p_internet = len(df[df['internet'] == 'yes'])/len(df)
p_internet
```

```
se = np.sqrt(0.90 * (1 - 0.90) / len(df)) # calculating standard error

be = p_internet # best estimate 
he =  0.90 # hypothesized estimate
test_stat = (be - he)/se # calculating test statistics

test_stat
```

```
pvalue = 2 * dist.norm.cdf(-np.abs(test_stat)) # calculating p-value

pvalue
```

```
if pvalue < 0.05:    # alpha value is 0.05 or 5%
   print("we are rejecting null hypothesis")
else:
  print("we are accepting null hypothesis")
```
